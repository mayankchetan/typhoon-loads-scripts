```python
tightCouple = "/projects/storm/IdlingRotor/decScratchRuns/testingBeforeFull/tightCoupling/23JAN24/steady"
tightCouple_Un = "/projects/storm/IdlingRotor/decScratchRuns/testingBeforeFull/tightCoupling/23JAN24/unsteady"

tightCouple_fast = "/projects/storm/IdlingRotor/decScratchRuns/testingBeforeFull/tightCoupling/23JAN24-Faster/steady"
tightCouple_fast_Un = "/projects/storm/IdlingRotor/decScratchRuns/testingBeforeFull/tightCoupling/23JAN24-Faster/unsteady"

ofv351 = "/projects/storm/IdlingRotor/decScratchRuns/testingBeforeFull/tightCoupling/23JAN24-of-351/steady"
ofv351_Un = "/projects/storm/IdlingRotor/decScratchRuns/testingBeforeFull/tightCoupling/23JAN24-of-351/unsteady"


# list of all the directories
dir_list = [tightCouple,tightCouple_Un,tightCouple_fast,tightCouple_fast_Un,ofv351,ofv351_Un]
```

```python
# Loop through all the directories, read the output files, and then compress them into dataframe and save them as csv

for dir in dir_list:
    # list the ".out" files in the directory
    files = [f for f in os.listdir(dir) if f.endswith(".out")]
    # loop through the files
    for file in files:
        # read the output file
        df = pd.read_csv(dir+"/"+file,skiprows=6,header=[0,1],dtype=np.float32,delim_whitespace=True)
        df.to_pickle(dir+"/"+file+".pkl")

```
